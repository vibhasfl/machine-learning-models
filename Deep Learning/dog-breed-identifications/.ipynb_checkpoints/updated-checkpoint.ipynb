```python
import os
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import tensorflow_hub as hub
from tensorflow.keras import layers, models
import tf_keras
from sklearn.model_selection import train_test_split
```

```python
labels_csv_df = pd.read_csv("dog-breed-data/labels.csv")
labels_csv_df.head(10)
```

```python
unique_breeds = sorted(labels_csv_df["breed"].unique().tolist())
```

```python
labels_csv_df["breed_index"] = labels_csv_df["breed"].map({breed: idx for idx, breed in enumerate(unique_breeds)})
```

```python
labels_csv_df.head(10)
```

#### Split data into train and validation sets

```python
train_df, val_df = train_test_split(labels_csv_df, test_size=0.2, stratify=labels_csv_df["breed_index"], random_state=42)
```

#### Load Image labels

```python
IMAGE_FOLDER = "dog-breed-data/train"
IMAGE_SIZE = (224,224)
def load_image(image_id,label):
    image_path = tf.strings.join([IMAGE_FOLDER, "/", image_id, ".jpg"])
    image = tf.io.read_file(image_path)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, IMAGE_SIZE)
    image = image/255.0
    return image, label

# load_image("0021f9ceb3235effd7fcde7f7538ed62",49)
```

```python
train_image_ids = train_df["id"].values
train_labels = train_df["breed_index"].values

val_image_ids = val_df["id"].values
val_labels = val_df["breed_index"].values
```

#### Create TF dataset

```python
train_dataset = tf.data.Dataset.from_tensor_slices((train_image_ids, train_labels)) # Still raw data
val_dataset = tf.data.Dataset.from_tensor_slices((val_image_ids, val_labels)) # Still raw data
```

```python
train_dataset = train_dataset.map(load_image, num_parallel_calls=tf.data.AUTOTUNE) # Actual data loading
val_dataset = val_dataset.map(load_image, num_parallel_calls=tf.data.AUTOTUNE) # Actual data loading
```

```python
BATCH_SIZE = 32

train_dataset = train_dataset.shuffle(buffer_size=1000)
train_dataset = train_dataset.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)


val_dataset = val_dataset.shuffle(buffer_size=1000)
val_dataset = val_dataset.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
```

#### Build model

```python
MODEL_URL = "https://tfhub.dev/google/imagenet/mobilenet_v2_100_224/classification/4"

feature_extractor_layer = hub.KerasLayer(
    MODEL_URL,
    input_shape=(224, 224, 3),
    trainable=False  # freeze weights
)
```

```python
num_classes = labels_csv_df["breed"].nunique()

model = tf_keras.Sequential([
    feature_extractor_layer,
    tf_keras.layers.Dense(units=num_classes, activation='softmax')  # your dog breed classes
])
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

#### Train model

```python
history = model.fit(train_dataset,validation_data=val_dataset, epochs=10)  # Adjust epochs as needed
```

```python
# Plot training and validation accuracy/loss
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs_range = range(len(acc))

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Train Accuracy')
plt.plot(epochs_range, val_acc, label='Val Accuracy')
plt.legend(loc='lower right')
plt.title('Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Train Loss')
plt.plot(epochs_range, val_loss, label='Val Loss')
plt.legend(loc='upper right')
plt.title('Loss')

plt.show()
```

```python
test_loss, test_acc = model.evaluate(val_dataset)
print(f"Test accuracy: {test_acc}")
```

```python
model.save('models/dog_breed_model_test_valid.keras')
```
